```python
#importing necessary library files
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
plt.style.use("fivethirtyeight")
```

```python
#reading csv file into dataframe
df=pd.read_csv("cvd.csv")
df.shape
```

```python
df["Deaths"]=df["Deaths"].replace("Deaths",0.0)
df["Deaths"]=df["Deaths"].astype("int64")
df["Confirmed"]=df["Confirmed"].replace("Confirmed",0.0)
df["Confirmed"]=df["Confirmed"].astype("int64")
```

```python
df.head()
```

```python
df=df
df["Confirmed_per_day"]=df["Confirmed_per_day"].replace("Confirmed_per_day",0)
df["Deaths_per_day"]=df["Deaths_per_day"].replace("Deaths_per_day",0)
df["Deaths_per_day"]=df["Deaths_per_day"].astype("int64")
```

```python
df["Total_Test_Results"].value_counts()
df["Total_Test_Results"]=df["Total_Test_Results"].replace("Total_Test_Results",0)
df["Total_Test_Results"]=df["Total_Test_Results"].astype("float64")
```

# Percentage of Missing values in Columns

```python
nulldetail=df.isnull().sum()
missing=nulldetail.to_frame().sort_values(by=0,ascending=False)
plt.style.use("bmh")
plt.figure(figsize=(8,10))
explode = [ 0, 0, 0.05, 0.06, 0.08, 0.1,0/1]
plt.title("Max Missing Columns",bbox={'facecolor':'0.9','pad':1})
plt.xticks()
plt.pie(missing[0].iloc[0:7],autopct='%.0f%%',shadow=True,labels=missing.index[0:7],explode=explode)
plt.show()
```

```python
#dropped people hospitalized as almost 80% of data is missing
df.drop(['People_Hospitalized','People_Tested'],axis=1,inplace=True)
```

# How each states were affected due to COVID-19 in the year 2021?

```python
df["Last_Update"]=df["Last_Update"].replace("Last_Update",0)
df["Confirmed_per_day"]=df["Confirmed_per_day"].astype("int64")
df["Last_Update"]=df["Last_Update"].astype("datetime64[ns]")
df["Year"] = [i.year for i in df["Last_Update"]]
stateAffected=df[df["Year"]==2021]
stateAffected=stateAffected.groupby("Province_State")["Confirmed_per_day","Deaths_per_day"].sum()
stateAffected=stateAffected.reset_index()
stateAffected=stateAffected.sort_values(by=["Confirmed_per_day","Deaths_per_day"],ascending=False)
stateAffected=stateAffected.head(15)
stateAffected
stateAffected.rename(columns = {'Deaths_per_day':'Deaths', 'Confirmed_per_day':'Confirmed'}, inplace = True)
```

```python
plt.style.use("fivethirtyeight")
stateAffected.plot(x="Province_State",y=["Deaths","Confirmed"],kind='bar')
plt.xlabel("States")
plt.ylabel("Confirmed Cases In 2021")
plt.title("Affect Due to COVID-19 in 2021")
plt.xticks(rotation=90)
plt.show()
```

# How was California affected in the month of January?

```python
df["Month"] = [i.month for i in df["Last_Update"]]
df["Province_State"]=df["Province_State"].astype("str")
calAffected1=df[(df["Month"]==1) & (df["Province_State"]=="California")]
calAffected=calAffected1.groupby("Last_Update")["Confirmed_per_day"].sum()
calAffected=calAffected.reset_index()
calDeathAffected=calAffected1.groupby("Last_Update")["Deaths_per_day"].sum()
calDeathAffected=calDeathAffected.reset_index()
calDeathAffected['Last_Update']=pd.to_datetime(calDeathAffected['Last_Update'],format='%Y-%m-%d %H:%M:%S')
calDeathAffected['Year3']=pd.DatetimeIndex(calDeathAffected['Last_Update']).year
calDeathAffected.drop(calDeathAffected[calDeathAffected['Year3']==2022].index, inplace = True)
calDeathAffected
calAffected['Last_Update']=pd.to_datetime(calAffected['Last_Update'],format='%Y-%m-%d %H:%M:%S')
calAffected['Year2']=pd.DatetimeIndex(calAffected['Last_Update']).year
calAffected.drop(calAffected[calAffected['Year2']==2022].index, inplace = True)
calAffected

calAffected
```

```python
plt.figure(figsize=(15,8))
plt.xlabel("January Month")
plt.ylabel("Counts")
plt.title("California Affected In the month of January")
sns.lineplot(x=calAffected["Last_Update"],y=calAffected["Confirmed_per_day"],label="Confirmed",palette='Set1',linewidth=3)
sns.lineplot(x=calDeathAffected["Last_Update"],y=calDeathAffected["Deaths_per_day"],label="Deaths",palette='Set2',linewidth=3)
plt.xticks(rotation=90)
plt.show()
```

```python
fig = plt.figure(figsize =(20, 20))
Date= calDeathAffected['Last_Update']
Death_Count = calDeathAffected['Deaths_per_day']
# Horizontal Bar Plot
plt.bar(Date,Death_Count,color ='pink')
plt.xticks(rotation=90)
plt.xlabel('Date', fontweight ='bold', fontsize = 15)
plt.ylabel('Deaths per day', fontweight ='bold', fontsize = 15)
plt.title("Deaths per day in California per day in January")
# Show Plot
plt.show()
```

# How was US affected due to COVID-19 in the course of two years(2020-21)? 

```python
a=df.groupby("Year")["Confirmed_per_day","Deaths_per_day","Total_Test_Results"].sum()
a=a.reset_index()
a=a.iloc[1:3]
a.rename(columns = {'Deaths_per_day':'Deaths', 'Confirmed_per_day':'Confirmed'}, inplace = True)
a
```

```python
a.groupby(['Year']).sum().plot(kind='pie', y='Deaths', autopct='%1.0f%%',startangle=100,title="Deaths in 2020-21  in USA")
```

```python
a.groupby(['Year']).sum().plot(kind='pie', y='Confirmed', autopct='%1.0f%%',startangle=100,title="Confirmed Cases in US 2020-21")
```

```python
a.rename(columns = {'Deaths_per_day':'Deaths', 'Confirmed_per_day':'Confirmed'}, inplace = True)
plt.style.use("fivethirtyeight")
plt.rcParams["figure.figsize"] = [15,8]
plots=a.plot(x="Year",y=["Confirmed","Deaths"],kind="bar",width=0.5)
plt.title("Affect of covid-19 in USA in 2020-21")
plt.ylabel("Counts")
for bar in plots.patches:
   plots.annotate(format(bar.get_height(), '.2f'),
                   (bar.get_x() + bar.get_width() / 2,
                    bar.get_height()), ha='center', va='center',
                   size=10, xytext=(0, 8),
                   textcoords='offset points')
```
